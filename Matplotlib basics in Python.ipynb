# Matplotlib basics in Python

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
##plotting using matplotlib

##plt scatter

plt.scatter(x,y, c='r')
plt.xlabel('X-axis')
plt.ylabel('Y-label')
plt.title('2D-graph')
plt.savefig('Test.png')
```

```python
y = x*x
```

```python
##plt.plot

plt.plot(x,y,'r*--',linewidth=2, markersize=12)
plt.xlabel('X-axis')
plt.ylabel('Y-label')
plt.title('2D diagram')
```

```python
##plotting subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'b')
```

```python
##Plotting a straight line

x = np.arange(1,11)
y = 3*x+10
plt.plot(x,y,'r*--', linewidth=2, markersize=12)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
##plotting sine and cosine curves in different subplots using the same x-values 

#setting up the values
x = np.arange(0, 4*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

#set up a subplot grid that has a height 2 and width 1
plt.subplot(2,1,1)

#create the sin plot
plt.plot(x,y_sin,'r')
plt.title('Sine plot')

#set up the 2nd subplot grid
plt.subplot(2,1,2)

#create the cosine plot
plt.plot(x,y_cos,'g')
plt.title('Cosine plot')
```

```python
##bar plots

x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data, vert=True, patch_artist=True)
```
