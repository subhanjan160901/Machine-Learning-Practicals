# Distribution Plots

1. **Distplot**

2. **Pairplot**

3. **Jointplot**

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

# Correlation with heatmap

A correlation heatmap uses colour cells, typically in a monochromatic scale, to show a 2D correlation matrix between various features. It is very important in Feature Selection.

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

# Jointplot

A jointplot allows us to study the relationship between two numeric variables. The central chart displays the correlation. It is usually a scatterplot, a hexbin, a 2D histogram or a 2D density plot.

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

# Pairplot

Pairplot shows all variables paired with other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

# Distplot

Distplot helps us to check the distribution of the columns feature in terms of histogram.

```python
sns.distplot(df['tip'])
```

Here the column section refers to the density. If we want to see the counts, we have to assign the 'kde' value as False.

**kde : Kernel Density estimation**

```python
sns.distplot(df['tip'], kde=False, bins=10)
```
